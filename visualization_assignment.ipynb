# GRAFICO 1 - PASAJEROS POR MES DEL AÑO 1949
## Line Plot
Lo elegí porque se puede ver como aumentan los pasajeros en junio y julio que es verano 
en el hemisferio norte


```python
import seaborn as datos
import matplotlib.pyplot as plt

# Cargar datos
base = datos.load_dataset("flights")

#seleccionar los primeros doce registros que son de 1949
base_1949 = base.head(12)

datos.lineplot(data=base_1949, x="month", y="passengers", marker="o", color="blue")

plt.title("Pasajeros por mes año: 1949")
plt.xlabel("Mes")
plt.ylabel("Cantidad de pasajeros")

# Guardar archivo
plt.savefig('line_plot_flight.png', dpi=150, bbox_inches='tight')
```

# GRAFICO 2 - Titanic - Supervivencia por clase
## Bar chart
Lo Elegí porque muestra que los mas pobres fueron los que menos sobrevivieron

```python
import seaborn as datos
import matplotlib.pyplot as plt

# Cargar datos
base = datos.load_dataset("titanic")

datos.barplot(data=base, x="class", y="survived", color="red")

plt.title("Supervivencia por clase")
plt.xlabel("Clase")
plt.ylabel("Cantidad de sobrevivientes")

# Guardar archivo
plt.savefig('barplot_titanic.png', dpi=150, bbox_inches='tight')
```

# GRAFICO 3 - Titanic - Grafico de Sobrevivientes
## Pie Chart
Este grafico muestra que sobrevivio el 38.4% del total de la gente

```python
import seaborn as datos
import matplotlib.pyplot as plt

# Cargar datos
base = datos.load_dataset("titanic")

# Contar sobrevivientes
contador = base["survived"].value_counts().sort_index()

etiqueta = ["No sobrevivió", "Sobrevivió"]

# Pie chart
plt.pie(contador, labels=etiqueta, autopct="%1.1f%%")
plt.title("sobrevivientes del Titanic")

# Guardar archivo
plt.savefig('piechart_titanic.png', dpi=150, bbox_inches='tight')

```

# GRAFICO 4 - Pingüinos y su peso
## Histograma
Lo Elegí porque muestra la distribución del peso de los pingüinos

```python
import seaborn as datos
import matplotlib.pyplot as plt

# Cargar datos
base = datos.load_dataset("penguins")

datos.histplot(data=base, x="body_mass_g", bins=30)

plt.title("Peso de pingüinos")
plt.xlabel("Peso")
plt.ylabel("Cantidad de Pinguinos")

# Guardar archivo
plt.savefig('histograma_pinguinos.png', dpi=150, bbox_inches='tight')
```

# GRAFICO 5 - Análisis de la cuenta de restaurant y la propina
## Scatter plot
Lo elegí porque en promedio la gente deja entre el 10% y 20% de la cuenta como propina y eso no depende de si es fumador o no

```python
import seaborn as datos
import matplotlib.pyplot as plt

# Cargar datos
base = datos.load_dataset("tips")

datos.scatterplot(data=base, x="total_bill", y="tip", hue="smoker")

plt.title("Análisis de cuenta y propina")
plt.xlabel("cuenta")
plt.ylabel("Propina")

# Guardar archivo
plt.savefig('scatterplot_propina.png', dpi=150, bbox_inches='tight')
```
